```python
plt.figure(figsize=(12, 8))

#Iteracija kroz kategorička polja i prikaz distribucije njihovih vrijednosti
for col in CATEGORIC_FIELDS:
    plt.figure(figsize=(8, 4))
    sns.histplot(german_credit_data[col],  bins=20,kde=True, color='green', edgecolor='black')
    plt.title(f'Histogram i distribucija za {col}', fontsize=12)
    plt.xlabel(col)
    plt.ylabel('Broj kategorija')
    plt.show()
    #print(category_lookup[col])
    for key, value in category_lookup[col].items():
        print(f"{key}: {value}")
```

<h4>Izračun srednje vrijednosti, z vrijednosti, zaobljenost distribucije, nagnutost distribucije</h4>

<p>Spremanje statistike u objekt "statistika_szzn" kroz koji onda prolazimo for petljom radi lakšeg ispisa</p>

```python
srednja_vrijednost = german_credit_data[NUMERIC_FIELDS].mean()
z_vrijednost = german_credit_data[NUMERIC_FIELDS].apply(zscore)
zaobljenost_distribucije =  german_credit_data[NUMERIC_FIELDS].apply(kurtosis)
nagnutost_distribucije = german_credit_data[NUMERIC_FIELDS].apply(skew)

statistika_szzn = {
    "Srednja vrijednost" : srednja_vrijednost,
    "Z vrijednosti": z_vrijednost,
    "Zaobljenost distribucije": zaobljenost_distribucije,
    "Nagnutost distribucije": nagnutost_distribucije,
}
for key, value in statistika_szzn.items():
    print(f"{key}:\n")
    print(value)
    print("-------------------------------------------\n")
```

<h3>Izračun korelacije svake varijable sa ciljnom varijablom: targetClass</h3> 


```python
korelacija = german_credit_data[NUMERIC_FIELDS].corr()[target_col]
#korelacija = korelacija.drop(column_names[20])
print(korelacija)
```

<h3>Izračun korelacije svake varijable izražene kroz Z vrijednost i  kategorizirane u razrede <-3 , (-3 do 3),>3 s ciljnom varijablom: targetClass</h3>

```python
def categorizeZ(z):
    if z < -3:
        return -3
    elif z > 3:
        return 3
    else:
        return 0

z_columns_names = NUMERIC_FIELDS[:]
z_columns_names.remove(column_names[20])
sve_z_vrijednosti = german_credit_data[z_columns_names].apply(zscore)
sve_z_vrijednosti_kategorije = sve_z_vrijednosti.map(categorizeZ)

german_credit_data_kategorizirano = pd.concat([german_credit_data[[target_col]], sve_z_vrijednosti_kategorije], axis=1)
korelacijaZ = german_credit_data_kategorizirano.corr()[target_col].drop(target_col)

print("Korelacije kategorizacijom\n")
print(korelacijaZ)
```

<h1>Zadatak 3</h1>

```python
def prediktivnost(IV):
    if(IV < 0.02):
        return "Varijabla nije korisna"
    elif(0.02 <= IV <= 0.1):
        return "Slaba prediktivnost"
    elif(0.1 < IV <= 0.3):
        return "Srednja prediktivnost"
    elif(0.3 < IV <= 0.5):
        return "Jaka prediktivnost"
    else:
        return "Suspicious (data leak?)"


german_data_copy = german_credit_data.copy()

num_copy = NUMERIC_FIELDS[:]
num_copy.remove(target_col) # ukljanjanje ciljne varijable

for col in num_copy:
    german_data_copy[col] = pd.qcut(german_credit_data[col], 20, duplicates="drop")

columns = column_names[:-1]

IV_german = pd.DataFrame()
IV_Data = []

for col in columns:
    a = german_data_copy[col]
    b = german_data_copy[target_col]

    IV = pd.crosstab(a,b).apply(lambda c: c/c.sum(), axis=0)

    good = IV[0]
    bad = IV[1]

    IV['WoE'] = np.log(good/bad)
    
    IV['IV'] = (good-bad) * IV['WoE']
    
    total_iv = IV['IV'].sum()

    woe_german = pd.DataFrame({'WoE': IV['WoE'], 'IV': IV['IV'], 'Prediktivnost': IV['IV'].apply(prediktivnost)}).round(4)
    woe_german.index.name = col
    IV_Data.append(woe_german)

    IV_german.loc[col, 'IV_suma'] = total_iv


#Ispis tablično svega
for data in IV_Data:
    print(f"\n{data.index.name} WoE tablica:")
    print(f"IV suma: {IV_german.loc[data.index.name,"IV_suma"]:.3f}")
    print(data.to_markdown(numalign="right", stralign="right"))

 
# Prikzaz srednje i jako prediktivne
IV_german_good = IV_german.sort_values('IV_suma', ascending=False).query('IV_suma > 0.1')

top_cols = IV_german_good.index
for col in top_cols:
    idx = columns.index(col)
    IV_Data[idx]['WoE'].plot(kind='bar', title=f'{col} (IV={IV_german.loc[col,"IV_suma"]:.3f})')
    plt.axhline(0, color='red', linestyle='--')
    plt.xticks(rotation=45)
    plt.show()
```
