<h1>Zadatak 3</h1>

```python
def prediktivnost(IV):
    if(IV < 0.02):
        return "Varijabla nije korisna"
    elif(0.02 <= IV <= 0.1):
        return "Slaba prediktivnost"
    elif(0.1 < IV <= 0.3):
        return "Srednja prediktivnost"
    elif(0.3 < IV <= 0.5):
        return "Jaka prediktivnost"
    else:
        return "Suspicious (data leak?)"


german_data_copy = german_credit_data.copy()

num_copy = NUMERIC_FIELDS[:]
num_copy.remove(target_col) # ukljanjanje ciljne varijable

for col in num_copy:
    german_data_copy[col] = pd.qcut(german_credit_data[col], 7, duplicates="drop")

columns = column_names[:-1].copy()
columns.remove(target_col_name)

IV_german = pd.DataFrame()
IV_Data = []

for col in columns:
    a = german_data_copy[col]
    b = german_data_copy[target_col]

    IV = pd.crosstab(a,b).apply(lambda c: c/c.sum(), axis=0)

    good = IV[0]
    bad = IV[1]

    IV['WoE'] = np.log(good/bad)
    
    IV['IV'] = (good-bad) * IV['WoE']
    
    total_iv = IV['IV'].sum()

    woe_german = pd.DataFrame({'Da': IV[0], 'Ne': IV[1], 'WoE': IV['WoE'], 'IV': IV['IV'], 'Prediktivnost': IV['IV'].apply(prediktivnost)}).round(4)
    woe_german.index.name = col
    IV_Data.append(woe_german)

    IV_german.loc[col, 'IV_suma'] = total_iv


#Ispis tablično svega
for data in IV_Data:
    print(f"\n{data.index.name} WoE tablica:")
    print(f"IV suma: {IV_german.loc[data.index.name,"IV_suma"]:.3f}")
    print(data.to_markdown(numalign="right", stralign="right"))

print()
```

```python
# Prikaz srednje i jako prediktivne
IV_german_good = IV_german.sort_values('IV_suma', ascending=False).query('IV_suma > 0.1')

top_cols = IV_german_good.index
for col in top_cols:
    idx = columns.index(col)
    IV_Data[idx]['WoE'].plot(kind='bar', title=f'{col} (IV={IV_german.loc[col,"IV_suma"]:.3f})')
    plt.axhline(0, color='red', linestyle='--')
    plt.xticks(rotation=45)
    plt.show()
```

<h3>Grupiranje WoE za sve varijable</h3>

```python
def woe_groups(df):
    """Grupira binove po WoE razini"""
    conditions = [
        (df['WoE'] > 0.6, 'Strong Positive'),
        ((df['WoE'] > 0) & (df['WoE'] <= 0.6), 'Weak Positive'), 
        ((df['WoE'] >= -0.6) & (df['WoE'] <= 0), 'Weak Negative'),
        (df['WoE'] < -0.6, 'Strong Negative')
    ]
    df['WoE_group'] = np.select([cond[0] for cond in conditions], 
                               [cond[1] for cond in conditions], 
                               default='Neutral')
    return df

IV_Data_grouped = []
for col in IV_Data:
    grouped = woe_groups(col.copy())
    summary = grouped.groupby('WoE_group').agg({
        'Da': 'sum', 'Ne': 'sum', 
        'WoE': 'mean', 'IV': 'sum'
    }).round(3)
    print(f"\n{col.index.name}:")
    print(summary.to_markdown())

print("\n")
```

<h3>Grupiranje razreda po kriteriju WoE za najjače varijable</h3>

```python
najjace_vars = IV_german[IV_german['IV_suma'] > 0.1].index.tolist()
print(f"Najjače varijable (IV > 0.1): {najjace_vars}")

for col in IV_Data:
    if col.index.name not in najjace_vars:
        continue
        
    grouped = woe_groups(col.copy())
    summary = grouped.groupby('WoE_group').agg({
        'Da': 'sum', 'Ne': 'sum', 'WoE': 'mean', 'IV': 'sum'
    }).round(3)
    
    print(f"\n{col.index.name} (IV={IV_german.loc[col.index.name, 'IV_suma']:.3f}):")
    print(summary.to_markdown())
```
